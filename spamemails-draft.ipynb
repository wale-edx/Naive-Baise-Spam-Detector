```python
def prepare_train_data(train=True):
    with open("data/train") as f:
        emails_train = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line
    train_size = len(emails_train)
    spam=0;ham=0
    spam_emails =[]
    ham_emails=[]
    for email in emails_train :
        attr=email.split(" ")
        if attr[1]== "spam":
            spam += 1
            spam_emails+=(attr[2:])
        else:
            ham +=1
            ham_emails+=(attr[2:])
    pspam=spam/ float(train_size)
    pham=ham/float(train_size)
    print "P(spam) : ",pspam
    print "P(ham) : " ,pham
    
    return spam_emails,ham_emails,pspam,pham
def get_word_dic(emails):
    word_dic={}
    e=np.array(emails)
    e=e.reshape(len(emails)/2,2)
    for i,x in e:
        try:
            word_dic[i]=(word_dic[i])+int(x)
        except:

            word_dic[i]=int(x)
    return word_dic

def top_five(dic):
    top=sorted(dic.iteritems(), key=lambda (v,k): (k,v),reverse=True)[:5]  
    return top
def get_dist_count(spam,ham):
    k={}
    for i in spam:
        try:
            k[i]=1
        except:
            k[i]=1
    for i in ham:
        try:
            k[i]=1
        except:
            k[i]=1
    return len(k)
def p_of_word(word,dic,k,name="ham",smooth=True):
    total_word = sum(dic.values())
    total_word = len(dic)
    try:
        p=(dic[word]+1)/(float(total_word)+k+1)
    except:
        p=1/(float(total_word)+k+1)
    #print "P(",word,"|",name,") = " ,p
    return p
def get_all_pr(dic,k,name="ham"):
    total_word = sum(dic.values())
    t=0
    for word in dic:
        p=p_of_word(word,dic,k)
        t=t+p
        print "P(",word,"|",name,") = " ,(p)
    return t
def NB(email,spam,ham,prior_spam,prior_ham):
    pins=prior_spam
    for word,count in email.items():
        pins=pins*(count)*NB_word(word,spam)
    pinh=prior_ham
    for word,count in email.items():
        pinh=pinh*(count)*NB_word(word,ham)
        
    if pins > pinh:
        return "spam"
    return "ham"
def NB_word(word,dic):
    return p_of_word(word,dic,1000)
    
```

```python
def prepare(emails):
    spam=0;ham=0
    labels=[]
    emails_list=[]
    for email in emails :
        label = email.split(" ")[1]
        labels.append(label)
        email=email.split(" ")[2:]
        email = get_word_dic(email)
        #email=np.array(email)
        #email=email.reshape(len(email)/2,2)
        emails_list.append(email)
    return labels,emails_list
```

```python
with open("data/test") as f:
    emails = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line

labels_test,emails_test=prepare(emails)
len_test= len(labels_test)
```
